```python
#Import required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from joblib import dump
from sklearn.metrics import accuracy_score
```

```python
#Load dataset
dataset = pd.read_csv(r'C:\Users\ASUS TUF\Downloads\archive (1)\Crop_Recommendation.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['Crop'].value_counts()
```

```python
x = dataset.drop(columns='Crop', axis = 1)
y = dataset['Crop']
```

```python
print(x)
```

```python
print (y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
classifier = RandomForestClassifier(random_state=42)
```

```python
classifier.fit(x_train, y_train)
```

```python
x_train_pred = classifier.predict(x_train)
trainning_acc = accuracy_score(x_train_pred, y_train)
```

```python
print("Akurasi Random Forest:", trainning_acc)
```

```python
x_test_pred = classifier.predict(x_test)
test_acc = accuracy_score(x_test_pred, y_test)
```

```python
print("Akurasi Random Forest:", test_acc)
```
